```python
#Store input requests
p_dict = {"AHMEDABAD":0, "AMRELI":1, "ANAND":2, "BHARUCH":3, "BHAVNAGAR":4 ,"BHUJ":5, "BOTAD":6, "DAHOD":7, "DEESA":8, "GANDHIDHAM":9, "GANDHINAGAR":10, "GODHRA":11, "GONDAL":12, "JAMNAGAR":13, "JETPUR":14, "JUNAGADH":15, "KALOL":16, "MEHSANA":17, "MORBI":18, "NADIAD":19, "NAVSARI":20, "PALANPUR":21, "PATAN":22, "PORBANDAR":23, "RAJKOT":24, "SURAT":25, "SURENDRANAGAR":26, "VADODARA": 27, "VALSAD":28, "VAPI":29, "VERAVAL":30}
fread = open("order.txt", "r")
lines = fread.readlines()
ship = []
for line in lines:
    t_h = line.split('\t') #change to \t if tab spaced
    print(t_h)
    src = p_dict[t_h[0]]
    dst = p_dict[t_h[1]]
    trucks = int(t_h[2])
    due = int(t_h[3].strip())
    ship.append([src, dst, trucks, due])
print(ship) 
n = len(ship)
n_ship = n
print(n)    
fread.close()
```

```python
#Calculate latest start & transit periods
transit = [0 for i in range(n)]
l_start = [0 for i in range(n)]
l_start_a = [0 for i in range(n)]
for i in range(n):
    src = ship[i][0]
    dst = ship[i][1]
    transit[i] = dist[src][dst]/50/24
    l_start[i] = ship[i][3] - transit[i] - 1
    l_start_a[i] = l_start[i] - dist[0][src]/60/24
print(transit)
print(l_start)
print(l_start_a)
```

```python
#Calculate the return dates to all source nodes
ret = [[0 for i in range(n)] for j in range(n)]
for i in range(n):
    src = ship[i][1]
    for j in range(n):
        dst = ship[j][0]
        ret[i][j] = ship[i][3] + 1 + dist[src][dst]/60/24
print(ret)
```

```python
#Creation of flow graph
g_list = []
#from ahm to all sources
for i in range(n):
    g_list.append([0, 2*i+1, 0, ship[i][2]])
#shipment edges
for i in range(n):
    g_list.append([2*i+1, 2*(i+1), 0, ship[i][2]])
#from dest to ahm
for i in range(n):
    g_list.append([2*(i+1), 2*n+1, 0, ship[i][2]])
#intermediate edges
for i in range(n):
    for j in range(n):
        if (j == i): #same shipment
            continue
        if (l_start[j] >= ret[i][j]): #go from i to j 
            g_list.append([2*(i+1), 2*j+1, 0, min(ship[i][2], ship[j][2])])
             
print(g_list)
```

```python
#(G, A}, Adjancency list & Capacity matrix
V = 2*(n+1)
E = len(g_list)
V_list = list(range(0,V,1))
A_dict = {}
G_mat = [ [ 0 for i in range(4) ] for j in range(E) ]
g = 0
for i in range(E):
    G_mat[g] = g_list[g]
    g += 1
for i in range(E):
    tail = G_mat[i][0]
    head = G_mat[i][1]
    if tail in A_dict.keys():
        A_dict[tail].append(head)        
    else:
        A_dict[tail] = [head]

capacity = [[0 for i in range(V)] for j in range(V)]
for i in range(E):
    tail = G_mat[i][0]
    head = G_mat[i][1]
    capacity[tail][head] = G_mat[i][3]

print(G_mat, "\n")
print(A_dict)   
print(capacity)
```

```python
def cap_Dijkstra(A_dict, cost, V, s, t):
    parent = {}
    for i in range(V):
        parent[i] = i
    dist = [math.inf]*V
    dist[s] = 0
    fixed = [s]
    chosen = s
    while (t not in fixed):
        if (chosen not in A_dict.keys()):
            print("cant use DJ here")
            return dist[t], parent
        chosen_range = A_dict[chosen]
        k = -1
        for i in range(len(chosen_range)):
            j = chosen_range[i]
            if (j in fixed):
                continue
            if (dist[j] >= dist[chosen] + cost[chosen][j]):
                dist[j] = dist[chosen] + cost[chosen][j]
                parent[j] = chosen
        min_d = 1e8
        min_v = -1
        for v in range(V):
            if ((dist[v] <= min_d) and (v not in fixed) and (dist[v] != math.inf) and (v in A_dict.keys() or v == t)):
                min_d = dist[v]
                min_v = v
        k = min_v
        if (k == -1 or k == math.inf):
            break
        else:
            fixed.append(k)
            chosen = k
    return dist[t], parent
```

```python
def create_residual_network(G_mat, A_dict, capacity):
    R = copy.deepcopy(G_mat)
    for i in A_dict.keys():
        for j in A_dict[i]:
            if (j in A_dict.keys()):
                if (i in A_dict[j]):
                    continue
            R.append([j, i, 0, 0])
    return R

# Develop modified dictionary
def mod_dict(R, capacity):
    R_dict = {}
    for e in range(len(R)):
        t = R[e][0]
        h = R[e][1]
        if (capacity[t][h] != 0):
            if (t in R_dict.keys()):
                if (h not in R_dict[t]):
                    R_dict[t].append(h)
            else:
                R_dict[t] = [h]
        else:
            if (t in R_dict.keys()):
                if (h in R_dict[t]):
                    R_dict[t].remove(h)
    return R_dict


def find_s_t_path(R, s, end, V, capacity): # Use R for nodes & arc connects and 'capacity' for residual capacity
    mod_cost = [[math.inf for i in range(V)] for j in range(V)]
    R_dict = {}
    
    #print('capacity')
    #for i in range(V):
    #    print(capacity[i])
        
    # Develop modified cost & dictionary
    for e in range(len(R)):
        t = R[e][0]
        h = R[e][1]
        if (capacity[t][h] != 0):
            mod_cost[t][h] = 1
            if (t in R_dict.keys()):
                if (h not in R_dict[t]):
                    R_dict[t].append(h)
            else:
                R_dict[t] = [h]
        else:
            if (t in R_dict.keys()):
                if (h in R_dict[t]):
                    R_dict[t].remove(h)
                    
    #Print values
    #print('modified cost')
    #for i in range(V):
    #    print(mod_cost[i])
    #print(R_dict)
    
    # Find shortest s-t path
    min_cost, parent = cap_Dijkstra(R_dict, mod_cost, V, s, end)
    #print(min_cost)
    #print(parent)
    path = []
    if (min_cost == math.inf):
        return path
    path = [end]
    curr = end
    while (curr != s and curr != parent[curr]):
        curr = parent[curr]
        path.insert(0, curr)
    #print(path)   
    return path

def retrieve_flow(G_mat, R, capacity):
    flow = []
    for i in range(len(G_mat)):
        t = G_mat[i][0]
        h = G_mat[i][1]
        old = G_mat[i][3]
        new = capacity[t][h]
        if (old > new):
            flow.append(old - new)
            #print("flow of", old-new, "from", t, "to", h)
        else:
            flow.append(0)            
    return flow

def FF(G_mat, A_dict, capacity, V, s, end):
    R = create_residual_network(G_mat, A_dict, capacity)
    #print(R)
    flow_value = 0
    path = find_s_t_path(R, s, end, V, capacity)
    #print(path)
    while (len(path) != 0):
        min_cap = math.inf
        for n in range(len(path)-1):
            t = path[n]
            h = path[n+1]
            if (capacity[t][h] < min_cap):
                min_cap = capacity[t][h]
        #print('here')
        #print(min_cap)
        flow_value = flow_value + min_cap
        for n in range(len(path)-1):
            t = path[n]
            h = path[n+1]
            capacity[t][h] = capacity[t][h] - min_cap
            capacity[h][t] = capacity[h][t] + min_cap
        path = find_s_t_path(R, s, end, V, capacity)
        #print(path)
    flow = retrieve_flow(G_mat, R, capacity)
    #print('flow')
    #print(flow)
    return flow, flow_value
```

```python
def back_to_depot(F_mat, n_ship, esrc, edst, etrucks, edue, estart, er_mat):
    t_satisfy = 0
    p_satisfy = 0
    t_ship = n_ship + len(er_mat)
    n_curr = 2*(t_ship)
    itr = len(F_mat)
    for i in range(itr):
        if (F_mat[i][1] == 2*n_ship+1):
            num = int(F_mat[i][0]/2) - 1
            if (num <= n_ship):
                src = ship[num][0]
            else: 
                e_count = F_mat[i][0] - 2*(n_ship + 1)
                src = er_mat[e_count][1] #change to actual source vaLue- using er_mat
            nstart = F_mat[i][2] + 1 + dist[src][esrc]/60/24
            if (nstart <= estart):
                p_satisfy = t_satisfy
                t_satisfy = min(p_satisfy + F_mat[i][4], etrucks)
                F_mat.append([F_mat[i][0], n_curr, F_mat[i][2], estart, t_satisfy - p_satisfy])
                if (t_satisfy < etrucks):
                    F_mat[i][4] = 0
                else:
                    t_satisfy = etrucks
                    F_mat[i][4] = F_mat[i][4] - t_satisfy
                    break
    t_remain = etrucks - t_satisfy
    if (t_satisfy != 0): #Satisfied by back to depot trucks
        F_mat.append([n_curr+1, 2*n_ship + 1, edue+1, edue+1+dist[edst][0]/60/24, t_satisfy])
    return t_remain
    

def free_shipment(F_mat, n_ship, esrc, edst, etrucks, edue, estart, er_mat):
    t_satisfy = 0
    p_satisfy = 0
    t_ship = n_ship + len(er_mat)
    n_curr = 2*(t_ship)
    itr = len(F_mat)
    for i in range(itr):
        if (F_mat[i][0]%2 == 0 and F_mat[i][1]%2 == 1):
            #print("Chosen Flow:")
            #print(F_mat[i])
            if (F_mat[i][0] >= 2*(n_ship + 1) or F_mat[i][0] == 0 or F_mat[i][1] == 2*n_ship+1): #ER shipments
                continue
            num = int(F_mat[i][0]/2)-1
            src = ship[num][1]
            nstart = F_mat[i][2] + 1 + dist[src][esrc]/60/24
            num = int((F_mat[i][1]-1)/2)
            #print(num)
            dst = ship[num][0]
            eend = edue + 1 + dist[edst][dst]/60/24
            nend = int(F_mat[i][3])
            if (nstart <= estart and eend <= nend): #start & end conditions satisfied
                p_satisfy = t_satisfy
                t_satisfy = min(p_satisfy + F_mat[i][4], etrucks)
                F_mat.append([F_mat[i][0], n_curr, F_mat[i][2], estart, t_satisfy - p_satisfy])  
                F_mat.append([n_curr + 1, F_mat[i][1], edue, F_mat[i][3], t_satisfy - p_satisfy])
                if (t_satisfy < etrucks):
                    F_mat[i][4] = 0
                else:
                    t_satisfy = etrucks
                    F_mat[i][4] = F_mat[i][4] - t_satisfy
                    break
    t_remain = etrucks - t_satisfy
    return t_remain            
                
def depot_to_src(F_mat, n_ship, esrc, edst, t_remain, edue, estart, er_mat):
    t_satisfy = 0
    p_satisfy = 0
    t_ship = n_ship + len(er_mat)
    n_curr = 2*(t_ship)
    itr = len(F_mat)
    for i in range(itr):
        if (F_mat[i][0] == 0):
            src = 0
            num = int((F_mat[i][1]-1)/2)
            dst = ship[num][0]
            eend = edue + 1 + dist[edst][dst]/60/24
            nend = int(F_mat[i][3])
            if (eend <= nend): #start & end conditions satisfied
                p_satisfy = t_satisfy
                t_satisfy = min(p_satisfy + F_mat[i][4], etrucks)
                F_mat.append([n_curr + 1, F_mat[i][1], edue, F_mat[i][3], t_satisfy - p_satisfy])
                if (t_satisfy < etrucks):
                    F_mat[i][4] = 0
                else:
                    t_satisfy = etrucks
                    F_mat[i][4] = F_mat[i][4] - t_satisfy
                    break
    if (t_satisfy != 0): #Satisfied by back to depot trucks
        F_mat.append([F_mat[i][0], n_curr, F_mat[i][2], estart, t_satisfy - p_satisfy]) 
    t_remain = etrucks - t_satisfy
    return t_remain

def start_depot(F_mat, n_ship, esrc, edst, t_remain, edue, estart, er_mat):
    t_satisfy = 0
    p_satisfy = 0
    t_ship = n_ship + len(er_mat)
    n_curr = 2*(t_ship)
    # depot to esrc
    nstart = e_start - dist[0][esrc]/60/24
    F_mat.append([0, n_curr, nstart, estart, t_remain])
    # edst to depot
    nend = edue + 1 + dist[0][edst]/60/24
    F_mat.append([n_curr+1, 2*n_ship+1, edue, nend, t_remain])     

            
```
